#### Making Description Categories into their own columns

```python
Descriptions = csv.Description.unique()
years = list(csv.columns[4:])

#QA
assert len(Descriptions) == 24, 'Looks like the number of descriptions changed. if so, this will affect the columns on the dataset on CIM'
#Visual Sanity Check.
Descriptions, years
```

```python
#Creating Geofips to GeoName map
geofips = csv.GeoFIPS.unique()
geoname = csv.GeoName.unique()

geofips_name = {key:value for key,value in zip(geofips, geoname)}
```

```python
geofips_name
```

####Actual Transform

```python
#Main Transform
final_df = pd.DataFrame()

for year in years:
  csv2 = pd.DataFrame(csv.groupby('GeoFIPS')[year].apply(list)) #list of expenditures in description field ordered in a list by geofips in a given year batch.
  csv2 = csv2[year].apply(pd.Series) #converts to a dataframe columns / #pd.explode should work too

  csv2.columns = Descriptions #new columns previously categorical col 'Descriptions'

  csv2['Year'] = year  #creating cat col for each year of rows
  columns = list(csv2.columns)
  columns.remove('Year')
  columns.insert(0,'Year') #Adding year to front of df
  csv2 = csv2[columns]

  final_df = final_df.append(csv2) # for each year batch append to final df. 

final_df.reset_index(inplace=True)
final_df.sort_values(['GeoFIPS','Year'],inplace = True)
final_df

#for future reference: alternatively, could've also: looped through each group, make group a dataframe with one column, append each group df to a main df, pd.explode the list column, finally change column name.
```

```python
#Mapping GeoName and Region
final_df['GeoName'] = final_df.GeoFIPS.apply(lambda x: geofips_name.get(x))
final_df['Region'] = final_df.GeoName.apply(lambda x: region_map.get(x))
```

```python
#reordering columns to CIM's format
columns = list(final_df.columns)
columns.remove('GeoName')
columns.remove('Region')
columns.insert(1,'Region') #Adding year to front of df
columns.insert(1,'GeoName')

final_df = final_df[columns]
```
