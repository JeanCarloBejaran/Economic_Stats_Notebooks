## Removing footers and fluff



```python
df = Full_data[Full_data['GeoName'].notna()].copy()
df.GeoName.value_counts()
#35 statistics per MSA
```

```python
df.query('Description == "Addenda:"').head(3)
```

##Transforms



```python
df.drop(columns = ['LineCode'], inplace = True)
```

```python
#testing there are 34 industries.
df = df[df.Description != 'Addenda:'].copy()
assert df.Description.nunique() == 34, 'There are more industries than expected' 
```

```python
#testing years column start hasn't shifted.
years = df.columns[3:]
try:
  [int(year) for year in years]
except:
  raise AssertionError ('Something is wrong with the years')
```

```python
#melting years columns into one with static GeoFips, Geoname, Description columns
final_df = df.melt(id_vars = ['GeoFips', 'GeoName', 'Description'], value_vars = years, var_name = 'Year', value_name = 'Thousands_dollars')
#Sort by Geofips, then Description, then year.
final_df.sort_values(['GeoFips', 'Description', 'Year'])
```

```python
#Breaking the '(D)' values into their own column and leaving a '' in its place

Blank_Key = pd.Series(data = ['' for row in final_df.Thousands_dollars], name = 'Blank_Key')
Thousands_dollars = final_df.Thousands_dollars.copy()
for id, entry in enumerate(final_df.Thousands_dollars):
  try:
    int(entry)
    Blank_Key.iloc[id] = ''
    Thousands_dollars.iloc[id] = entry
  except:
    Blank_Key.iloc[id] = entry
    Thousands_dollars.iloc[id] = ''
```

```python
final_df['Thousands_dollars'] = Thousands_dollars
final_df['Blank_Key'] = Blank_Key
```

```python
#cleaner if desired.
#final_df = filtered(final_df)
```

###Renaming columns

```python
final_df.rename(columns = {'GeoName' : 'msa_name', 'Description':'Industry'}, inplace = True )
```

###Cleaning msa name to match CIM format:

```python
final_df['msa_name'] =  final_df.msa_name.apply(lambda x: x.replace(' (Metropolitan Statistical Area)', ''))
final_df['msa_name'] = final_df.msa_name.apply(lambda x: x.replace(' *', ''))
```
